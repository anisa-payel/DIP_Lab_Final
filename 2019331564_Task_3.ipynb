```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/MyDrive/DIP Images/DIP_Lab_task_03.png'
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if image is None:
    print("Error: Image not found. Check the path.")
else:
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    unique, counts = np.unique(image, return_counts=True)
    intensity_distribution = dict(zip(unique, counts))
    print("Intensity Distribution:", intensity_distribution)
    print("Possible Noise:")
    if 0 in intensity_distribution or 255 in intensity_distribution:
        print("Detected possible salt-and-pepper noise.")
    else:
        print("No significant salt-and-pepper noise detected.")

    kernel_size = (5, 5)
    smoothed_image = cv2.GaussianBlur(image, kernel_size, 0)

    plt.subplot(1, 2, 2)
    plt.title("Smoothed Image")
    plt.imshow(smoothed_image, cmap='gray')
    plt.axis('off')
    plt.show()
```
