```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/MyDrive/DIP Images/DIP_Lab_task_01.png'
```

```python
image = cv2.imread(image_path)
if image is None:
    print("Error: Image not found. Check the path.")
else:
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    image1 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    height, width = image1.shape
    print(f"Image Dimensions: Width = {width}, Height = {height}")

    min_intensity = np.min(image1)
    max_intensity = np.max(image1)
    print(f"Intensity Range: Min = {min_intensity}, Max = {max_intensity}")

    intensity_counts = cv2.calcHist([image1], [0], None, [256], [0, 256])

    plt.figure(figsize=(8, 6))
    plt.title("Intensity Frequency Distribution")
    plt.xlabel("Intensity Value")
    plt.ylabel("Frequency")
    plt.plot(intensity_counts, color='black')
    plt.grid()
    plt.show()
```
