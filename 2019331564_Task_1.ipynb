```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
image_path = '/content/DIP_Lab_task_01.png'
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if image is None:
    print("Error: Image not found. Check the path.")
else:
    height, width = image.shape
    print(f"Image Dimensions: Width = {width}, Height = {height}")

    min_intensity = np.min(image)
    max_intensity = np.max(image)
    print(f"Intensity Range: Min = {min_intensity}, Max = {max_intensity}")

    intensity_counts = cv2.calcHist([image], [0], None, [256], [0, 256])

    plt.figure(figsize=(8, 6))
    plt.title("Intensity Frequency Distribution")
    plt.xlabel("Intensity Value")
    plt.ylabel("Frequency")
    plt.plot(intensity_counts, color='black')
    plt.grid()
    plt.show()
```
