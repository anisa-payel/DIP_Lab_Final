**Task_4 :** Implement an appropriate technique to reduce the brightness of the provided image (DIP_Lab_task_04.jpg) effectively. [20]

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/MyDrive/DIP Images/DIP_Lab_task_04.jpg'
```

```python
image = cv2.imread(image_path)
if image is None:
    print("Error: Image not found. Check the path.")
else:
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    h, s, v = cv2.split(hsv_image)
    brightness_scale = 0.5
    v = np.clip(v * brightness_scale, 0, 255).astype(np.uint8)
    adjusted_hsv = cv2.merge([h, s, v])
    adjusted_image = cv2.cvtColor(adjusted_hsv, cv2.COLOR_HSV2BGR)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.title("Brightness Reduced Image")
    plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
