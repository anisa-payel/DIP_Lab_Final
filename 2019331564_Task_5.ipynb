```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
image_path = '/content/drive/MyDrive/DIP Images/DIP_Lab_task_05.jpg'
```

```python
image = cv2.imread(image_path)
if image is None:
    print("Error: Image not found. Check the path.")
else:
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    threshold_value, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

    kernel = np.ones((3, 3), np.uint8)
    refined_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Binary Segmentation")
    plt.imshow(binary_image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Refined Segmentation")
    plt.imshow(refined_image, cmap='gray')
    plt.axis('off')

    plt.show()
```
